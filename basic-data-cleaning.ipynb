```python
## importing required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
## reading the example csv -- Heart Disease UCI (Kaggle)
df = pd.read_csv("csv/heart.csv")
df.head()
```

```python
# dropping a set of unused columns
to_drop = ['cp','fbs','restecg','thalach','exang','oldpeak','slope','thal','target','ca']
#drop set, specifying inplace to rename the date in place (returning nothing) and axis=1 for columns
df.drop(to_drop, inplace=True, axis=1)
df.head()
```

```python
# rename columns for convenience
# store old and new values in a dictionary
new_name = {'age': 'Age', 'sex': 'Sex', 'trestbps': 'Bps', 'chol': 'Cholesterol'}
# replace old values (new_name keys) with new values (new_name values)
df.rename(columns = new_name, inplace = True)
df.head()
```

```python
# recode Sex values to M and F
replace_values = {0: "F", 1: "M"}
# apply encoding dictionary with .replace
df = df.replace({"Sex": replace_values})
df.head()
```

### Now I'll apply this to a dataset of my choosing:

Medaling Men's Hammer Throw results from the Modern Olympics

```python
athletics = pd.read_csv("csv/results.csv")
hdf = athletics.loc[athletics['Event'] == 'Hammer Throw Men']
```

```python
hdf.dtypes
```

```python
hdf['Result'] = hdf['Result'].astype('float')
pd.DataFrame(hdf)
```

Let's say I just want to create a visualization which shows what results earned which medal in each Olympic Games.

```python
drops = ['Gender','Event','Location','Name','Nationality','Wind']

hdf = hdf.drop(drops, axis=1)
```

```python
hdf.head()
```

```python
# change medal to a score (3 points for gold, 2 for silver, 1 for bronze)
new_name = {'Year': 'Games', 'Medal': 'Score', 'Result': 'Mark'}

hdf.rename(columns = new_name, inplace=True)
hdf.head()
```

```python
medal_vals = {"G":3, "S":2, "B":1}
hdf = hdf.replace({"Score": medal_vals})
hdf.head()
```

```python
hdf.groupby("Games")
hdf = hdf.sort_values(by="Games")
X, y = hdf['Games'], hdf['Mark']
z = hdf['Score']

hdf.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
fig = plt.figure()
ax = plt.axes()

import matplotlib as mpl
cmap = mpl.colors.LinearSegmentedColormap.from_list("", ["darkgoldenrod","silver","gold"])

ax.scatter(X, y, c=z, cmap=cmap)
ax.set_title("Men's Hammer Throw Medaling Performances (Modern Olympics)")
ax.set_xlabel("Year")
ax.set_ylabel("Result(m)");
```
