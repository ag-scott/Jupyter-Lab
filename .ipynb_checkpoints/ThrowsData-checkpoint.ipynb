```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import sklearn
from sklearn import datasets
```

```python
results = pd.read_csv('results.csv')
```

```python
results['Event'].unique()
```

```python
if np.any(results['Event'].str.contains('Men')):
    results['Gender'] = 'M'
else:
    results['Gender'] = 'F'
    
results.head()
```

```python
event = results['Event']

results['Event Group'] = 0

results['Event Group'][event.str.contains("Discus|Shot Put|Javelin|Hammer")] = "Throws"
results['Event Group'][event.str.contains("Jump|Vault")] = "Jumps"
results['Event Group'][event.str.contains("100M|200M|400M Men|400M Women")] = "Sprints"
results['Event Group'][event.str.contains("800M|1500M|3000M")] = "Mid-Distance"
results['Event Group'][event.str.contains("5000M|10000M|20Km|50Km|Marathon")] = "Long-Distance"
results['Event Group'][event.str.contains("Relay")] = "Relay"
results['Event Group'][event.str.contains("Decathlon|Heptathlon")] = "Multi"

results.head()
```

```python
throws = results.loc[results['Event Group'] == 'Throws']
throws.head()
```

```python
throws.shape
```

```python
menshot = throws.loc[throws['Event'] == 'Shot Put Men']
```

```python
menshot.sort_values(by='Year')
```

```python
menshot['Result'] = menshot['Result'].astype('float')
```

```python
menshot['Result'].dtype
```

```python
olyrec = menshot.loc[menshot['Result'] == 22.52]
```

```python
hfont = {'fontname':'Times New Roman'} 

plt.style.use('fivethirtyeight')

fig = plt.figure()
ax = plt.axes()

X = menshot['Year']
y = menshot['Result']
plt.scatter(X, y, facecolor='lightblue', edgecolor='k')

plt.scatter(olyrec['Year'], olyrec['Result'], facecolor='gold', edgecolor='k', marker='*', s=120, alpha=1)

ax.set_xlabel('Year', **hfont)
ax.set_ylabel('Result (m)', **hfont)
ax.set_title('Medaling Shot Put Performances, Olympics 1896-Present', **hfont)
```

```python
menshot['4y'] = menshot['Result'].rolling(4).mean()
```

```python
menshotcopy = menshot.copy()
menshotcopy.sort_values(by='Year')
```

```python
menshotcopy = menshotcopy.set_index('Year')
menshotcopy['4y'] = menshotcopy['Result'].rolling(4).mean()
```

```python
menshotcopy.groupby('Year')
```

```python
menshotcopy = menshotcopy.sort_values(by='Year')
```

```python
menshotcopy['4y'] = menshotcopy['Result'].rolling(4).mean()
menshotcopy['8y'] = menshotcopy['Result'].rolling(8).mean()
menshotcopy['3y'] = menshotcopy['Result'].rolling(3).mean()
```

```python
menshotcopy['4y'].plot()
menshotcopy['8y'].plot()
menshotcopy['3y'].plot()

plt.legend()
```

```python
plt.style.use('fivethirtyeight')

fig = plt.figure()
ax = plt.axes()

X = menshot['Year']
y = menshot['Result']
plt.scatter(X, y, facecolor='lightblue', edgecolor='k', label='Marks', zorder=2)

plt.scatter(olyrec['Year'], olyrec['Result'], facecolor='gold', edgecolor='k', marker='*', s=120, alpha=1, label = 'Olympic Record', zorder=3)

plt.plot(menshotcopy['4y'], color='gray', linewidth=2, label='Moving Average', linestyle="--", zorder=1)

ax.set_xlabel('Year', **hfont)
ax.set_ylabel('Result (m)', **hfont)
ax.set_title('Medaling Shot Put Performances, Olympics 1896-Present', **hfont)
plt.legend()
```
